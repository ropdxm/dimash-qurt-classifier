```python
import requests

def search_images_google(query, api_key, cse_id, num_images=10):
    urls = []

    for start in range(1, num_images + 1, 10):
        url = "https://www.googleapis.com/customsearch/v1"

        params = {
            "key": api_key,
            "cx": cse_id,
            "q": query,
            "searchType": "image",
            "num": min(10, num_images - len(urls)),
            "start": start
        }

        response = requests.get(url, params=params)
        response.raise_for_status()
        results = response.json()

        if "items" not in results:
            break

        for item in results["items"]:
            urls.append(item["link"])

    return urls
```

```python
from pathlib import Path
from fastai.vision.all import download_images, resize_images

def download_images_google(query, folder, api_key, cse_id, num_images=50):
    folder = Path(folder)
    folder.mkdir(parents=True, exist_ok=True)

    urls = search_images_google(query, api_key, cse_id, num_images)

    download_images(folder, urls=urls)
    resize_images(folder, max_size=400)

    print(f"Downloaded {len(urls)} images.")
```
